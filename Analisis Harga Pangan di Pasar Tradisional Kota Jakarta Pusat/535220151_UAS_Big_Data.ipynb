```python
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
#harian
path ="Jakarta Pusat.xlsx"
df = pd.read_excel(path,sheet_name='Sheet2')
df.head()
df['Day']=df['Date'].dt.day_name()
```

```python
var = ['Beras Kualitas Super I', 'Daging Sapi Kualitas 1', 'Bawang Merah Ukuran Sedang', 'Minyak Goreng Kemasan Bermerk 2', 'Gula Pasir Kualitas Premium']
dfy1 = df.groupby(['Day']).mean() [var]
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dfy1 = dfy1.reindex(days)
dfy1.plot(kind='bar')
plt.xticks(range(len(dfy1.index)), dfy1.index)
plt.ylabel('harga')
plt.title('rata rata harga variabel')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.5), ncol=3)
plt.show()
```

```python
var = ['Beras Kualitas Super I', 'Daging Sapi Kualitas 1', 'Bawang Merah Ukuran Sedang', 'Minyak Goreng Kemasan Bermerk 2', 'Gula Pasir Kualitas Premium']
dfy1 = df.groupby(['Day']).max() [var]
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dfy1 = dfy1.reindex(days)
dfy1.plot(kind='bar')
plt.xticks(range(len(dfy1.index)), dfy1.index)
plt.ylabel('harga')
plt.title('maksimal harga variabel')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.5), ncol=3)
plt.show()
```

```python
var = ['Beras Kualitas Super I', 'Daging Sapi Kualitas 1', 'Bawang Merah Ukuran Sedang', 'Minyak Goreng Kemasan Bermerk 2', 'Gula Pasir Kualitas Premium']
dfy1 = df.groupby(['Day']).min() [var]
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dfy1 = dfy1.reindex(days)
dfy1.plot(kind='bar')
plt.xticks(range(len(dfy1.index)), dfy1.index)
plt.ylabel('harga')
plt.title('minimal harga variabel')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.5), ncol=3)
plt.show()
```
